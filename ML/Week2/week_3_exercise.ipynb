# Q2 Create 20 random numbers and create respective graphs

```python
import random as r
l1 = []
for a in range(20):
    l1.append(r.randint(1,100))
```

```python
plt.plot(l1)
plt.show()
```

```python
sns.histplot(l1)
plt.show()
```

# bar plot

```python
l2 = []
for x in range(20):
    l2.append(x)
sns.barplot(x= l2,y=l1)
plt.show()
```

# side by side graph

```python
a = pd.DataFrame({"1":l1,"2":l2})
a.plot(kind='bar')
plt.show()
```

# stacked graph

```python
b = pd.DataFrame({"1":l1,"2":l2})
b.plot(kind='bar',stacked=True)
plt.show()
```

# sin graph

```python
import math
x = np.arange(0,math.pi*2,0.05)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

# 3D scatter plot using seaborn and matplotlib

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')
sns.set_style("whitegrid")
ax.scatter(l1,l2)
ax.set_xlabel("X AXIS")
ax.set_ylabel("Y AXIS")
ax.set_zlabel("Z AXIS")
plt.show()
```

# surface plot

```python
l1 = []
l2 = []
l1.append(np.arange(-1,1,0.02))
l2.append(np.arange(-1,1,0.02))
l1,l2 = np.meshgrid(l1,l2)
l3 = l1**2+l2**2
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel("X AXIS")
ax.set_ylabel("Y AXIS")
ax.set_zlabel("Z AXIS")
ax.plot_surface(l1,l2,l3)
plt.show()
```

# Q5 Create a matrix using numpy and perform transpose multiplication and inverse of the matrix

```python
r = int(input("Enter number of rows"))
c = int(input("Enter number of columns"))
l1 = []
i = 0
while(i<r*c):
    l1.append(int(input()))
    i=i+1
arr = np.array([l1])
arr1 = np.matrix(arr.reshape(r,c))
arr1
```

```python
np.transpose(arr1)
```

```python
np.linalg.inv(arr1)
```

```python
arr2 = [[1,2],[3,4]]
res = np.dot(arr1,arr2)
res
```
